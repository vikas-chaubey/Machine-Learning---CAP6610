```python
from keras.datasets import mnist
import numpy as np
from sklearn.metrics import accuracy_score

def accuracy_percentage(actual_value: np.ndarray, predicted_value: np.ndarray):
    #print(actual_value.shape,"  ",predicted_value.shape)
    #count which outcomes are same
    matches = (actual_value == predicted_value).sum()
    #print("matches", matches)
    total_samples = predicted_value.shape[0]
    #print("total_samples", total_samples)
    percentage_accuracy = (matches/total_samples)*100
    print("percentage_accuracy when iteration=",i,percentage_accuracy)
    return percentage_accuracy




#loading trainng and testing data sets from mnist datset file
(training_set_x, training_set_y), (testing_set_x, testing_set_y) = mnist.load_data()

#creating vectors out of image matrices in the training set
training_set_x = training_set_x.reshape([training_set_x.shape[0], -1])

#creating vectors out of image matrices in the testing set
testing_set_x = testing_set_x.reshape([testing_set_x.shape[0], -1])

#in order to start the proximal gradient descent , parameters has to be initialized with some initial value
parameter_assumed = np.random.randn(784,10)

#length of training set
training_set_length = len(training_set_x)

#calculating predicted_outcome for training set by dot product with assumed parameters
predicted_outcome = training_set_x.dot(parameter_assumed)
yi_predicted_outcome = predicted_outcome[np.arange(predicted_outcome.shape[0]),training_set_y]

#obtaining the class_prediction function 
class_prediction_expression = np.maximum(0,predicted_outcome - np.matrix(yi_predicted_outcome).T + 1)
class_prediction_expression[np.arange(training_set_length),training_set_y] = class_prediction_expression.min()


#obtaining loss function
loss_function = np.mean(class_prediction_expression.max(1))
loss_function += 0.5 * 0.01 * np.sum(np.sqrt(np.sum(np.square(parameter_assumed),1)))

#initializing initial values of parameters with assumed values
parameter = parameter_assumed

#array to hold the accuracy obtained every 1000 iterations
accuracies = []


# for k in range(5):
iterations = 10**6
for i in range(iterations):
    t = i%60000
    #initializing parameters with 0 value
    dparameter = np.zeros(parameter.shape)

    #calculating predicted_outcome for training set by dot product with assumed parameters
    predicted_outcome  = training_set_x[t].dot(parameter)

    for j in range(10):
        if predicted_outcome[j]>0:
            dparameter[:,j] += training_set_x[t]

    #optimization of parameters value with a random image data 
    dparameter[:,training_set_y[t]]+= -1*training_set_x[t]
    regularizer_value = parameter/np.sqrt(np.sum(np.square(parameter),1)).reshape(784,1)
    parameter_assumednorm = np.linalg.norm(parameter - 0.05*dparameter - 0.01*regularizer_value)
    g = (0.01*0.05)/parameter_assumednorm
    parameter= (1-g)*(parameter - 0.05*dparameter - 0.01*regularizer_value)
    if i%1000 == 0:
        #prediction of results on training data set with currently optimized paramer values after completing 1000 iteration 
        predict = np.argmax(testing_set_x.dot(parameter),axis =1)
        accuracy_percentage(testing_set_y,predict)
```
